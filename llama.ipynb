```python
import torch
from langchain import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig, pipeline

MODEL_NAME = "TheBloke/Llama-2-13b-Chat-GPTQ"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME, torch_dtype=torch.float16, trust_remote_code=True, device_map="auto"
)

generation_config = GenerationConfig.from_pretrained(MODEL_NAME)
generation_config.max_new_tokens = 1024
generation_config.temperature = 0.0001
generation_config.top_p = 0.95
generation_config.do_sample = True
generation_config.repetition_penalty = 1.15

text_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    generation_config=generation_config,
)

llm = HuggingFacePipeline(pipeline=text_pipeline, model_kwargs={"temperature": 0.1})
```

```python
result = llm(
    "Give the answer  for the following question in the format [question]: Lets think step by step - [reason] Answer: [answer]. If Jack has 30 seashells. He gives some to Sally. Sally now has 21 seashells. How many seashells does Jack have?"
)
print(result)
```

```python
result = llm("What is 7 times 8 divided by 4")
print(result)
```

```python
from langchain import PromptTemplate, ChatPromptTemplate

template = """
<s>[INST] <<SYS>>
Answer in the format Lets think step by step followed by the reasoning and then the answer at the end.
<</SYS>>
{text} [/INST]
"""

# add_sub_prompt = PromptTemplate.from

prompt = PromptTemplate(
    input_variables=["text"],
    template=template,
)
```

```python
text = "If Jack has 30 seashells. He gives some to Sally. Sally now has 21 seashells. How many seashells does Jack have?"
print(prompt.format(text=text))
```

```python
result = llm(prompt.format(text="You throw a ball 5 feet away, and it comes right back to you. Which direction did you throw it in?"))
print(result)
```

```python
result = llm(prompt.format(text="How many months have 28 days"))
print(result)
```

```python
result = llm(prompt.format(text="3. Which is heavier: 100 pounds of steel or 100 pounds of feathers?"))
print(result)
```
